```python
import requests
from bs4 import BeautifulSoup
```

```python
web_site = requests.get('https://www.naver.com/')
soup = BeautifulSoup(web_site.content , 'html.parser')
```

```python
print(soup)
```

```python
all_divs = soup.find_all('div')
print(all_divs)
```

```python
first_div = soup.find('div')
print(first_div)
```

```python
div_id = soup.find('div', {'id' : 'gnb'})
print(div_id)
```

```python
# .json()으로 읽지 않고 데이터를 읽기를 시도.

read_test = requests.get('https://www.naver.com/srchrank?frm=main')
ranks_test = BeautifulSoup(read_test.content , 'html.parser')

print(ranks_test)
```

```python
# data를 읽을 때 ranks_test에 있는 content에서 data를 뽑아도 None값이 나온다.

ranks_data = ranks_test.get("data")
print(ranks_data)
```

```python
# find로 읽기 시도를 했으나, data는 tag가 아니라 읽을 수 없음. => .json() 으로 읽자.

ranks_data = ranks_test.find('data')
print(ranks_data)
```

```python
# 네이버 실검은 json(java script object notation) file로 읽어와서 보여주므로 실검 json파일 원본을 읽어서 data만 출력한다. 

json = requests.get('https://www.naver.com/srchrank?frm=main').json()
ranks = json.get('data')

print(ranks)
```

```python
for rank in ranks:
    # 각 데이터는 rank, keyword, keyword_synomyms
    
    number = rank.get("rank")
    keyword = rank.get("keyword")
    keyword_synonyms = rank.get("keyword_synonyms")
    
    print(str(number) + '.', keyword,', 동의어: ' +  str(keyword_synonyms))
    

    '''
    if keyword_synonyms == []:
        print(str(number) + '.', keyword)
    
    else:    
        print(str(number) + '.', keyword,'| 동의어: ' +  str(keyword_synonyms))
    '''
```
